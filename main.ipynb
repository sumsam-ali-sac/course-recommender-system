```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
print('Dependencies Imported')
```

```python
courses_data = pd.read_csv('Courses_data.csv')
courses_data.head()
```

```python

def clean_and_tokenize(text):
    text = text.lower().replace(',', ' ')
    text = text.lower()
    text = "".join([char for char in text if char not in string.punctuation])
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    return tokens

courses_data['processed_tags'] = courses_data['tags'].apply(clean_and_tokenize)
courses_data[['tags', 'processed_tags']].head()
```

```python
from transformers import TFBertModel, BertTokenizer
import tensorflow as tf
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np

# Load pre-trained BERT with TensorFlow
model_name = 'bert-base-uncased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = TFBertModel.from_pretrained(model_name)

def generate_embeddings(text):
    inputs = tokenizer(text, return_tensors='tf', padding=True, truncation=True, max_length=512)
    outputs = model(inputs)
    return outputs.last_hidden_state[:, 0, :].numpy()

courses_data['Embeddings'] = courses_data['processed_tags'].apply(generate_embeddings)
courses_data['Embeddings'] = courses_data['Embeddings'].apply(lambda x: x.flatten())

def recommend_courses(input_text, courses_data, top_n=5):
    input_embedding = generate_embeddings(input_text).flatten()
    course_embeddings = np.vstack(courses_data['Embeddings'])
    similarities = cosine_similarity([input_embedding], course_embeddings)[0]
    top_indices = similarities.argsort()[-top_n:][::-1]
    return courses_data.iloc[top_indices]

input_text = "advanced machine learning"
recommended_courses = recommend_courses(input_text, courses_data)
print(recommended_courses)
```

```python
def recommend_courses(input_text, df, top_n=5):
    input_embedding = generate_embeddings(input_text).flatten()
    course_embeddings = np.vstack(df['Embeddings'])
    similarities = cosine_similarity([input_embedding], course_embeddings)[0]
    top_indices = similarities.argsort()[-top_n:][::-1]
    return df.iloc[top_indices]

# Example usage
input_text = "python programming"
recommended_courses = recommend_courses(input_text, courses_data, top_n=5)
print(recommended_courses)
```
